# 函数的多返回值

```python
# 演示使用多个变量，接收函数的多个返回值
def test_return():
    return 1, 'hello', True
x, y, z = test_return()
print(x)
print(y)
print(z)
```

# 函数的多种传参方式
1. 位置参数
2. 关键字参数
3. 不定长参数
4. 缺省参数

```python
# 定义一个函数
def user_info(name, age, gender):
    print(f'姓名是:{name},年龄是:{age},性别是:{gender}')

# 位置参数——按函数定义的位置顺序进行传参-默认使用形式
user_info('小明',20,'男')

# 关键字参数——再传参时给出形参赋值即可（ps：顺序自由）
user_info(name='小王',age='21',gender='男')
user_info(age='22',name='小鹏',gender='男')    # 可以不按照参数的定义顺序传参
user_info('蔚来',gender='女',age=26)   # 关键字参数和位置参数可以混合使用

# 缺省参数
"""
用于定义函数，为参数提供默认值，调用函数时可不传入该默认参数的值。
ps：所有位置参数必须出现在默认参数前，包括函数定义和调用——即默认参数必须写在所有形参的最后面
作用：当调用函数时没有传递参数时，就会使用默认是缺省参数对应的值
当函数调用时，若缺省参数传入了值则修改默认参数值，否则就使用该默认值
"""
# 定义一个函数
def user_info(name, age, gender='女'):
    print(f'姓名是:{name},年龄是:{age},性别是:{gender}')
user_info('小天',13)
```

# 不定长参数
又称可变参数，用于不确定调用的时候会传递多少个参数（不传参也可以）的场景。
作用：当调用函数时不确定参数个数时，可以使用不定长参数
不定长参数的类型:1.位置不定长传递 2.关键词不定长传递


```python
# 1.位置不定长传递——*args
# ——不定长定义的形式参数会作为元组存在，接收不定长数量的参数传入
# 其中args代表：传入的所有参数都会被args变量收集，它会根据传入参数的位置合并为一个元组（tuple），args是元组类型，这就是位置传递
# 其中*代表：接收的参数数量是无限的（包含0）
def user_info(*args):
    print(args)
# ('Tom',)
user_info('Tom')
# ('Tom', 18)
user_info('Tom', 18)
```

```python
# 2.关键字不定长传递——**kwargs
# ——不定长定义的形式参数作为字典存在，接收“键=值”形式的传入参数，且数量不限。
# 其中kwargs代表：参数是“键=值”形式的情况下，所有的“键=值”都会被kwargs接收，同时会根据“键=值”组成字典（dict），kwargs是字典类型，这就是关键字传递
# 其中**代表：接收参数数量是无限的（包含0），且传入参数必须满足“键=值”的形式
def user_info(**kwargs):
    print(kwargs)
# {'name': 'Tom', 'age': 18, 'id': 110}
user_info(name='Tom', age=18,id = 110)
```

# 匿名函数
1. 函数作为参数传递
2. lambda匿名函数

```python
# 1.函数作为参数传递
"""
函数compute作为参数，传入了test_func函数中进行了使用
1.test_func需要一个函数作为参数传入，其内部的compute(1,2)就是函数调用的形式，因此其计算逻辑由被传入的函数决定
2.compute函数接收两个数字，并对其进行运算，compute函数作为参数，传递给了test_func函数使用
3.最终，再test_func函数内部，由传入的compute函数，完成了对数字计算的操作
总结：这是一种计算逻辑(函数)的传递，而非传统数据的传递
"""
def test_func(compute):
    result = compute(1,2)   # 确定compute是函数，此处为其的调用
    print(f'计算结果:{result}')
    print(f'传入参数compute的类型是:{type(compute)}')
    
def compute(x,y):
    return x+y

test_func(compute)  # 3
```

# lambda匿名函数
"""
函数定义中
·def关键字，可以定义带有名称的函数
·lambda关键字，可以定义匿名函数（无名称）
有名称的函数，可以基于名称重复使用
无名称的匿名函数，只可临时使用一次
"""

匿名函数的定义语法：
lambda 传入参数: 函数体(一行代码)
·lambda是关键字，表示定义匿名函数
·传入参数表示匿名函数的形式参数，如x,y表示接收2个形式参数
·函数体，就是函数的执行逻辑，要注意:只能写一行，无法写多行代码

```python
# lambda匿名函数用法

# 用def关键字定义函数
def test_func(compute):
    result = compute(1,2)
    print(result)

def compute(x,y):
    return x+y
test_func(compute)      # 输出：3

# 也可通过lambda关键字，传入一个一次性使用的lambda匿名函数
def test_func(compute):
    result = compute(1,2)   # 形参为函数
    return result
# 传入lambda匿名函数
test_func(lambda x,y:x+y)   # 输出：3
```
