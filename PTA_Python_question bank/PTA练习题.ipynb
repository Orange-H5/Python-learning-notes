# str.join()——join()方法
# 通用模板
保留的字符集合 = {"字符1", "字符2", ...}  
或直接用字符串 "字符1字符2..."
新字符串 = '分隔字符串'.join([字符 for 字符 in 原字符串 if 字符 in 保留的字符集合])

# 总结
join() 核心是「用指定分隔符拼接可迭代对象的字符串元素」，语法为 分隔符.join(可迭代对象)；
基础用法：空字符串拼接（无缝合并）、单 / 多字符拼接（格式化输出）；
关键避坑：非字符串元素必须先转字符串（用 str() + 列表推导式）；

# 例子
s = "abc123!@#456xyz789"
keep_chars = "0123456789"   //只保留 0-9 的数字字符
result = ''.join([c for c in s if c in keep_chars])
print(result)  # 输出：123456789

# 10进制转为16进制——hex(10进制数字)   eg：a = hex(16)  # 0x10
# 16进制转为10进制——int(16进制数字, base = 16)
若16进制数字之前有0x则可以省略base参数
# int() 是 Python 内置函数，支持将任意进制的字符串 / 数字转为十进制整数，是处理十六进制转十进制的首选。

```python
# 26
str1 = input()
# 第一次清洁，保留字符串
str_retain1 = '0123456789abcdefABCDEF-' 
str_retain2 = '0123456789abcdefABCDEF'  
# 通过拼接字符串，遍历str1，若遍历的str1元素存在于str——retain1中则存入str1_1
str1_1 = ''.join(str2 for str2 in str1 if str2 in str_retain1)
str1_2 = ''.join(str3 for str3 in str1 if str3 in str_retain2)
# 判断第一次清洁后的字符串str1_1是否为空
if str1_2 == '':
    result = 0
else:
    # 判断在第一个十六进制符前面是否含有-号
    if str1_1[0] == '-':
        result = '-'
    else:
        result = ''
    # 去除所有‘-’号
    result_str = str1_1.replace('-', '')
    result += str(int(result_str, 16))
print(int(result))
```

——Python 中 map() 方法的用法，首先明确核心：map() 是 Python 内置的高阶函数，用于将一个「函数」应用到一个 / 多个「可迭代对象」（列表、元组等）的每个元素上，返回一个迭代器（map 对象），核心作用是批量处理可迭代对象的元素，替代繁琐的循环。
——map(函数, 可迭代对象1[, 可迭代对象2, ...])
”“”
函数：可以是内置函数、自定义函数、lambda 匿名函数，要求能接收「可迭代对象元素数量」的参数（比如处理 2 个可迭代对象时，函数需接收 2 个参数）；
可迭代对象：1 个或多个（列表、元组、字符串等），多个时会按位置一一对应取元素；
返回值：map 迭代器（需转为列表 / 元组才能直观看到结果）。
”“”
eg：
# 需求：将字符串列表转为整数列表
str_list = ["1", "2", "3", "4"]
# map将int()应用到每个元素，返回迭代器
result_iter = map(int, str_list)
# 转为列表查看结果
result_list = list(result_iter)
print(result_list)  # 输出：[1, 2, 3, 4]

ps：
map()方法，返回值是迭代器：map() 返回的是 map 迭代器，不是列表 / 元组，需用 list()/tuple() 转换才能查看结果；且迭代器只能遍历一次，遍历后会清空：
iter_obj = map(int, ["1", "2"])
print(list(iter_obj))  # 输出：[1, 2]
print(list(iter_obj))  # 输出：[]（迭代器已空）

# 用map()方法接收多个输入值
——利用 map() 方法接收多个输入值，核心思路是：先通过 input() 获取一行输入的字符串，用 split() 拆分出多个值，再用 map() 将拆分后的字符串批量转换为目标类型（如 int/float），最后转为列表 / 元组提取具体值。这种方式是 Python 中接收多输入的高效写法，下面分场景详细讲解。
eg：
最常用的场景是用户在一行输入多个数字（如 1 2 3 4），需要将其转为整数 / 浮点数列表。
# 1. 接收输入并拆分 → 2. map转换类型 → 3. 转为列表/元组
输入值列表 = list(map(目标类型, input().split()))
1.input().split()：将输入的一行字符串按空格拆分（无论多少个空格都拆分），返回字符串列表；
2.map(目标类型, ...)：将拆分后的每个字符串转为目标类型（如 int/float）；
3.list(...)：将 map 迭代器转为列表，方便后续取值。
空格分隔的多值（通用写法）
适用于输入用「任意数量空格」分隔的场景（如 1 2 3、a b c），用 split() 拆分（默认按空格拆分，自动忽略多余空格）。
具体代码：

# enumerate()方法
enumerate(iterable, start=0)
参数：
iterable：必选，可迭代对象（列表、元组、字符串、字典、生成器等）。
start：可选，索引起始值，默认 0(可指定为 1、10 等任意整数)。
——此start并不代表列表中的真实索引值，而是通过该方法重新定义输出的起始索引值。即遍历仍从iterable[0]开始，只是另外定义输出的起始索引。参考eg2
返回值：返回一个枚举对象（迭代器），每次迭代输出一个（索引，元素）的元组
核心作用：遍历可迭代对象时，同时获取元素的索引和值

```python
# eg1：
# 遍历列表，同时获取索引和元素
lst = ['a', 'b', 'c']
for idx, val in enumerate(lst):
    print(f'索引：{idx},元素：{val}')
```

```python
# 核心用法1:指定索引起始值（start 参数）
# 适合「索引从 1 开始」的场景（如统计排名、行号）
lst = ['苹果', '香蕉', '橙子']
for idx, val in enumerate(lst, start = 1):
    print(f'第{idx}个水果：{val}')
```

```python
# 核心用法2:遍历字符串（字符 + 索引）
s = "Python"
for idx, char in enumerate(s):
    print(f'索引{idx}:{char}')
```

# dict.items()方法——字典方法
——dict.items()
参数：无参数
返回值：「字典视图对象（dict_items）」，包含字典中所有的 (键, 值) 元组
核心作用：一次性获取字典的「键」和「值」，而非单独获取键或值，是遍历字典键值对的首选方式。
总结：
1.核心作用：items() 是字典遍历键值对的首选方法，返回包含 (键, 值) 元组的视图对象；
2.经典搭配：和 enumerate() 配合可同时获取「索引 + 键 + 值」，需注意 (key, value) 的括号不能省；
3.关键区别：
items() → 键 + 值；keys() → 仅键；values() → 仅值；
视图对象是动态的，需固定则转为列表 / 元组；
避坑要点：遍历中修改字典键会报错，修改值则允许；解包时必须加括号。

```python
# eg1:基础用法（单独使用items()）
# 定义一个示例字典
student = {'name': 'Tom', 'age': 18, 'gender': 'male'}
# 调用items()获取键值对视图
key_value = student.items()
print(f'items()返回值类型：{type(key_value)}')
print(f'items()返回值：{key_value}')

# 遍历键值对（最常用方法）
for key, value in student.items():
    print(f'键:{key},值:{value}')
    
```

```python
# 核心用法1：和 enumerate() 配合（获取索引 + 键 + 值）
student = {'name': 'Tom', 'age': 18, 'gender': 'male'}
for idx, (key, value) in enumerate(student.items()):
    print(f'第{idx}组键值对 → 键：{key}，值：{value}')
```

```python
# 核心用法2：转换为列表 / 元组（可视化 / 持久化键值对）
# 字典视图对象（dict_items）是「动态的」（原字典修改后，视图会同步变化），若需固定键值对，可转为列表 / 元组：
student = {'name': 'Tom', 'age': 18}
# 转为列表
kv_list = list(student.items())
print(kv_list)

# 转为元组
kv_tuple = tuple(student.items())
print(kv_tuple)

# 字典视图对象的动态性：修改原字典，视图同步变化
student['gender'] = 'male'
print(f'已经固定为列表的视图{kv_list}')
print(f'已经固定为元组的视图{kv_tuple}')
print(f'为进行固定的原始视图student.items():{student.items()}')
```

# 对比字典的三个核心方法
dict.items()
返回值：所有（键，值）元组视图
作用：遍历键+值（最常用）
eg：for k, v in dict.itmes()

dict.keys()
返回值：所有键的视图
作用：仅遍历键
eg：for key in dict.keys()

dict.values()
返回值：所有值的视图
作用：仅遍历值

```python
# 常见误区
# 误区1：解包时省略括号
# 和enumerate()配合时，(key, value) 的括号不能省，否则会触发「值数量不匹配」错误：
student = {'name': 'Tom'}
# 错误写法：无括号，解包时把 ('name','Tom') 拆分为两个变量，enumerate返回的是 (0, ('name','Tom')) → 共2个元素，却要拆3个
# for idx, key, value in enumerate(student.items()):  # 报错：ValueError: not enough values to unpack (expected 3, got 2)

# 正确写法：加括号
for idx, (key, value) in enumerate(student.items()):
    print(idx, key, value)  # 输出：0 name Tom
    
# 误区2：修改视图对象
# items()返回的是「视图」，不是列表，不能直接修改（如添加 / 删除元素）：
d = {'a':1}
# d.items().append(('b',2))  # 报错：'dict_items' object has no attribute 'append'

# 正确做法：先转为列表，再修改
kv_list = list(d.items())
kv_list.append(('b',2))
print(kv_list)  # 输出：[('a', 1), ('b', 2)]

# 误区3：遍历过程中修改字典
# 遍历 items()时直接修改字典的「键」（如新增 / 删除键），会触发迭代错误；修改「值」则没问题：
d = {'a':1, 'b':2}

# 错误：遍历中删除键
# for k, v in d.items():
#     del d[k]  # 报错：RuntimeError: dictionary changed size during iteration

# 正确：先转列表，再遍历
for k, v in list(d.items()):
    del d[k]
print(d)  # 输出：{}

# 允许：遍历中修改值
for k, v in d.items():
    d[k] = v * 2
```

```python
# 79
"""
你需要用「一重循环 + 字典」实现两数之和问题（解唯一，找不到输出 no answer），核心思路是用字典记录已遍历元素的「值 - 索引」，遍历过程中检查「目标数 - 当前值」是否在字典中，这样只需一次循环就能找到结果，时间复杂度 O (n)。下面给出完整代码、逐行注释和核心逻辑讲解。
暴力双重循环：
1.遍历每个元素，再遍历后续元素找互补值
2.时间复杂度O(n²)
3.无需额外空间

一重循环＋字典：
1.遍历元素时，用字典已遍历元素的[值-索引]，直接查找互补值是否存在字典中
2.时间复杂度O(n)，效率更高
3.空间复杂度O(n),(字典存储）
"""
nums = list(map(int,input().split(',')))
target = int(input())
# 定义空字典，用于储存当前遍历数字v的互补数字n
dict_1 = {}
# 定义result，来存入结果，并用于后续判断是否存在唯一解
result = None
# 遍历nums列表中的索引值idx以及对应的元素值v
for idx, v in enumerate(nums):
    # v为目标数1，n为目标数2。要保证v+n=target。确定n
    # 即获取当前遍历数字v的互补数字n，若后续遍历碰到n则得到结果——因此要存入字典备用
    n = target - v
    # 若后续的遍历的v值存在于dict_1中，则说明有互补数字
    if n in dict_1:
        result = (dict_1[n], idx)
        break
    dict_1[v]=idx
if result:
    print(f'{result[0]} {result[1]}')
else:
    print('no answer')
```

```python
# 87
N = int(input())
# 接收输入
# 用该形式字典存入数据dict1 = {ID:[name,goal1,goal2,foal3]}
dict_ID_name = {}
for i in range(N):
    ID,name,goal1,goal2,goal3 = list(map(str,input().split()))
    dict_ID_name[ID]=[name,goal1,goal2,goal3]
# 因为该最高分学生是唯一的，所以最高成绩是唯一的，
# 因此可以构建字典{sum(goal1,goal2,goal3):ID}来确定该学生
dict_sum_goals = {}
for ID,list_name_goal in dict_ID_name.items():
    sum_goals = sum(map(int,list_name_goal[1::]))
    dict_sum_goals[sum_goals] = ID
# 找到总成绩最大值
max_goal = max(dict_sum_goals)
# 总成绩最大值对应的ID
max_ID = dict_sum_goals[max_goal]
# 总成绩最大值对应的name
max_name = dict_ID_name[max_ID][0]
# 输出结果
print(f'{max_name} {max_ID} {max_goal}')
```
